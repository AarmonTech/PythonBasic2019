```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 10, 500)


fig, ax = plt.subplots()

line1, = ax.plot(x, np.sin(x), '--', linewidth=2,
                 label='sine-Wave')

line2, = ax.plot(x, -1 * np.tan(x),
                 label='Dashes set proactively')



ax.legend(loc='lower right')
plt.show()
```
